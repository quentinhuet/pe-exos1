```python
import numpy as np
import matplotlib.pyplot as plt

square = np.ones((10,10),dtype=bool)
plt.imshow(square)
```

```python
def cadre(A):
    haut, long = np.shape(A)
    A[1,:]= 0
    A[haut-2,:]=0
    A[:,1]=0
    A[:,long-2]=0
```

```python
A = square.copy()
cadre(A)
plt.imshow(A)
```

```python
def cadre2(A):
    haut, long = np.shape(A)
    A[0,:]= 0
    A[haut-1,:]=0
    A[:,0]=0
    A[:,long-1]=0
```

```python
B = square.copy()
cadre2(B)
plt.imshow(B)
```

```python
list = [1,2,3,4]
```

```python
def anti_diagonal(list):
    n = len(list)
    res = np.empty((n,n))
    res[:,:] = 0
    for i in range(n):
        res[i,n-i-1] = list[i]
    print(res)
```

```python
anti_diagonal([1,2,5,9])
```
